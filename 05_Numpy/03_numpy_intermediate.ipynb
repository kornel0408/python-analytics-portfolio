```python
import numpy as np
```

```python
# creating array
subjects = ["Maths", "Science", "English"]
std_marks = np.array([
    [78, 85, 90],
    [67, 77, 82],
    [90, 91, 89],
    [56, 65, 70],
    [88, 79, 92]  
])

print("Subjects:",subjects)
print("Student_marks:", std_marks)
```

```python
shape = std_marks.shape
print("Shape of the table :", shape)
```

```python
d_type = std_marks.dtype
print("Data type:", d_type)
```

```python
table_dimension = std_marks.ndim
print(f"Dimesion of the table: {table_dimension} dimension")
```

```python
# accessing student1 maaths marks
math_marks = std_marks[0,0]
print("1st student Maths marks :",math_marks)

#accessing science marks of the 3rd student
s_mrks = std_marks[2,1]
print("3rd student science marks:", s_mrks)

#accessing english marks of the 5th student
e_marks = std_marks[4,2]
print("5th student english marks:", e_marks)
```

```python
#Accessing all subjects marks of all the student

maths_marks = std_marks[:,0]
science_marks = std_marks[:,1]
english_marks = std_marks[:,2]

print("Maths marks:")
print(maths_marks)

print("Science_marks")
print(science_marks)

print("English marks:")
print(english_marks)
```

```python
# Change student 4 English marks to 75
print("array before modification:", std_marks)
std_marks[4,2] = 75
print("array after modification", std_marks)
```

```python
# Change student 2 math to 70
# Change student 5 science to 80

print("array before modification:", std_marks)
std_marks[1,0] = 70
std_marks[4,1] = 75
print("array after modification", std_marks)
```

```python
# Add 5 marks to all
print("array before adding 5 marks:")
print(std_marks)

std_marks = std_marks + 5
print("array after adding 5 marks:")
print(std_marks)
```

```python
# Find average marks

avg_marks = std_marks.mean()
print(round(avg_marks))
```

```python
# Find max
# Find min
# Find sum

max_marks = std_marks.max()
print("Maximun marks:")
print(max_marks)

min_marks = std_marks.min()
print("Minimun marks")
print(min_marks)

total_marks = std_marks.sum()
print("Total marks")
print(total_marks)
```
