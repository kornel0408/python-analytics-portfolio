```python
# **************************************
# Dataset 1 — Sales Data (E-commerce)
# **************************************
```

```python
import numpy as np

sales = np.array([
    [101, 500, 10],
    [102, 300, 5],
    [103, 700, 8],
    [104, 200, 20],
    [105, 1000, 2]
])
```

```python
# Task 1: Basic Info
# Find shape
# Find dtype
# Number of products?
```

```python
print("Sahpe of the array:",sales.shape)
print("Data type:", sales.dtype)
print("Number of producs", len(sales))
```

```python
# Task 2: Revenue Calculation

# Revenue = Price × Quantity
# Create new array for revenue
# Print revenue for each product
```

```python
for i in range(len(sales)):
    print(f"product_id: {sales[i][0]} Revenue: {sales[i][1]*sales[i][2]}")
```

```python
# Task 3: Total Business Revenue
# Find total revenue using NumPy
```

```python
Revenue_sales = np.array(sales[:,1]*sales[:,2])
total_revenue = Revenue_sales.sum()
print("Total Revenue = ",total_revenue)
```

```python
# Task 4: Highest Selling Product
# Find product with highest quantity sold
```

```python
for i in range(len(sales)):
    if sales[i][2] == sales[:,2].max():
        high_qty = sales[i][0]
        print(f"Highest Quantity product= {high_qty} and Qty = {sales[:,2].max()}")
```

```python
# Task 5: Increase Price
# Increase all prices by 10%
```

```python
price_before_incresing = sales[:,1]
print("Product price:",price_before_incresing)

price_after_increasing = sales[:,1]+sales[:,1]*(10/100)
print("Price after increasing 10%:", price_after_increasing)
```

```python
# Task 6: Add New Product
# [106, 400, 15]
```

```python
new_product = np.array([[106, 400, 15]])
sales = np.append(sales, new_product, axis=0)
print(sales)
print(sales.shape)
```

```python
# 👉 Combine everything

# Create dataset:

# 10 products
# price between 100–1000
# quantity between 1–20

# Do:

# ✅ total revenue
# ✅ average price
# ✅ max quantity
# ✅ increase price
# ✅ add new product
# ✅ print final dataset
```

```python
product_id = np.arange(1,11).reshape(10,1)
product_price = np.random.randint(100, 1000,(10,1))
product_qty = np.random.randint(1,10,(10,1))

products = np.hstack((product_id,product_price,product_qty))
```

```python
print(products)
```

```python
# ✅ total revenue
total_revenue_products = products[:,1] * products[:,2]
products_total_r = total_revenue_products.sum()
print("Total Revenue =", products_total_r)

# ✅ average price
avg_price = products[:,1].mean()
print("Average price = ", avg_price)

# ✅ max quantity
max_qty = products[:,2].max()
print("Highest Quantity = ", max_qty)

# ✅ increase price
products[:,1] = products[:,1]+100
print(products[:,1])

# ✅ add new product
new_prod = np.array([[11, 300, 8]])
products = np.append(products,new_prod, axis=0)

# ✅ print final dataset
print("Final products:\n",products)
```
