```python
# **************************************
#     Dataset 1 — E-commerce Orders
# **************************************
```

```python
# columns = [Order_ID, Amount, Quantity, Discount%]
```

```python
import numpy as np

orders = np.array([
    [1001, 5000, 2, 10],
    [1002, 2000, 1, 5],
    [1003, 8000, 4, 20],
    [1004, 1500, 1, 0],
    [1005, 6000, 3, 15],
    [1006, 3000, 2, 0]
])
```

```python
# Task 1 — High Value Orders
# Find orders where:
# Amount > 5000
```

```python
high_value_orders = orders[orders[:,1]>5000]
print(high_value_orders)
```

```python
#Task 2 — Discount Analysis
# Find orders where:
# Discount > 10%
```

```python
high_discount = orders[orders[:,3]>10]
print(high_discount)
```

```python
# Task 3 — Multiple Condition
# Find orders where:
# Amount > 3000 AND Quantity >= 2
```

```python
mult_condtn = orders[(orders[:,1]>3000) & (orders[:,2]>=2)]
print(mult_condtn)
```

```python
# Task 4 — Apply Discoun
# Final Amount = Amount - Discount
# Modify dataset:
# Apply discount only where Discount > 0
```

```python
# Step 1: Create condition
condition = orders[:,3] > 0

# Step 2: Apply discount
orders[condition, 1] = orders[condition, 1] - (
    orders[condition, 1] * orders[condition, 3] / 100
)

print(orders)
```

```python
# Task 5 — Label Orders
# Create labels:
# "High" if Amount > 5000
# "Low" otherwise
```

```python
labels = np.where(orders[:,1]>5000, "High", "Low")
print(labels)
```

```python
# ****************************************
#     Dataset 2 — Employee Performance
# ****************************************  
```

```python
# columns = [Emp_ID, Salary, Experience, Rating]
```

```python
emp = np.array([
    [1, 25000, 1, 3],
    [2, 50000, 4, 5],
    [3, 70000, 6, 4],
    [4, 30000, 2, 2],
    [5, 90000, 8, 5],
    [6, 45000, 3, 3]
])
```

```python
# Task 1 — Top Performers
# Find employees where:
# Rating >= 4
```

```python
top_performers = emp[emp[:,3]>=4]
print(top_performers)
```

```python
# Task 2 — Senior High Earners
# Find employees:
# Experience > 5 AND Salary > 60000
```

```python
high_earners = emp[(emp[:,2]>5) & (emp[:,1]>6000)]
print(high_earners)
```

```python
# Task 3 — Salary Increment
# Increase salary by 10%:
# Only for employees with Rating >= 4
```

```python
condition2 = emp[:,3] >= 4

emp[condition2,1] = emp[condition2,1] + (
    emp[condition2,1] * emp[condition2,3] / 100
)

print(emp)
```

```python
# Task 4 — Create Category Column
# "Senior" → Experience > 5
# "Mid" → 3–5
# "Junior" → < 3
```

```python
experience = emp[:,2]

category = np.where(
    experience > 5, "Senior",
    np.where(experience >= 3, "Mid", "Junior")
)

print(category)
```

```python
# Task 5 — Replace Low Ratings
# If Rating < 3 → make it 3

ratings = emp[:,3] < 3
emp[ratings,3] = 3
print(emp)
```

```python
# **************************************************
#     Dataset 3 — Student Result Analysis
# **************************************************
```

```python
# [Math, Science, English]
marks = np.array([
    [78, 85, 90],
    [40, 35, 50],
    [88, 92, 95],
    [60, 65, 70],
    [30, 25, 40]
])
```

```python
# Task 1 — Failed Students
# Find students where:
# Any subject < 35
```

```python
failed_students = marks[np.any(marks<35, axis=1)]
print(failed_students)
```

```python
# Task 2 — Top Students
# Find students where:
# All subjects > 80
```

```python
top_students = marks[np.all(marks>80, axis=1)]
print(top_students)
```

```python
# Task 3 — Add Grace Marks
# Add 5 marks:
# Only to students with average < 50
```

```python
avg = np.all(marks<marks.mean(), axis=1)
marks[avg] = marks[avg] + 5
print(marks)
```

```python
# Task 4 — Pass/Fail Label
# "Pass" → all subjects >= 35
# "Fail" → otherwise
```

```python
result = np.where(np.all(marks>=35, axis=1), "pass", "fail")
print(result)
```

```python
# Task 5 — Subject-wise Analysis
# Find max marks per subject
# Find average marks per subject
```

```python
max_marks = np.max(marks, axis=1)
print("MAXIMUM MARKS:")
for i in range(len(max_marks)):
    print(f"Student{i+1} Maximum marks = {max_marks[i]}")
print("\n")

avg_marks = np.mean(marks, axis=1)
print("AVERAGE MARKS:")
for i in range(len(max_marks)):
    print(f"student{i+1} Average marks = {round(avg_marks[i])}")
```

```python
# Create dataset:

# 10 customers
# purchase amount (1000–10000)
# visits (1-10)

# Do:

# high value customers (>5000)
# frequent customers (>5 visits)
# give discount:

# 10% if both conditions true
# label:
# "Premium" / "Regular"
```

```python
purchase_amount = np.random.randint(1000, 10000,(10,1))
visits = np.random.randint(1,10,(10,1))
```

```python
customers = np.hstack((purchase_amount, visits))
print("Customer details:\n", customers)
```

```python
# high value customers (>5000)
high_value_customers = customers[customers[:,0]>5000]
print("High Value customers: \n", high_value_customers)
```

```python
# frequent customers (>5 visits)customers[customers[:,1]>5]
frequent_customers = customers[customers[:,1]>5]
print(frequent_customers)
```

```python
# give discount:
# 10% if both conditions true

discount_cndtn1 = customers[:,0]>5000
discount_cndtn2 = customers[:,1]>5
discount_cndtn = discount_cndtn1 & discount_cndtn2

customers[discount_cndtn, 0] = customers[discount_cndtn, 0] - (customers[discount_cndtn, 0] * 10/100)
print("After Discount: \n:", customers)

```

```python
# "Premium" / "Regular"
customer_type = np.where(discount_cndtn, "Premium", "Regular")
print(customer_type)
```
