```python
# Pandas = Data Analysis Library in Python

# It is used for:

# Working with tables (like Excel)
# Data cleaning
# Data analysis
# Handling missing values
# Filtering data
```

```python
# Why Pandas?

# You have data like:

# Name	Salary	Experience
# A	25000	1
# B	40000	3

# In NumPy → only numbers, no column names 
# In Pandas → structured table with names 
```

```python
import pandas as pd
```

```python
#1D data
salary = pd.Series([25000, 40000, 60000])
print(salary)
```

```python
# With Custom Index
salary = pd.Series([25000,40000,60000], index=["A","B","C"])
print(salary)
```

```python
# DataFrame (very important)

# Method : From Dictionary
data = {
    "Name": ["A", "B", "C"],
    "Salary": [25000, 40000, 60000],
    "Experience": [1, 3, 5]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Basic Info
print(df.shape)   # rows, columns
print(df.columns) # column names
print(df.dtypes)  # data types
```

```python
# Access Column
print(df["Salary"])
```

```python
# Add New Column
df["Bonus"] = [2000, 3000, 5000]
print(df)
```

```python
#Deleting column
df = df.drop("Name", axis = 1)
```

```python
print(df)
```

```python
#Adding constant column
df["PF"] = 1500
print(df)
```

```python
#checking data
first_2_rows = df.head(2)
print(first_2_rows)
```

```python
last_2_rows = df.tail(2)
print(last_2_rows)
```
