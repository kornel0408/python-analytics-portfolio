```python
# Create this dataset:

# Product	Price	Quantity
# A	        100	        10
# B      	200	        5
# C	        150	        8 
```

```python
import pandas as pd
```

```python
products = {
    "product" : ['A', 'B', 'C'],
    "Price" : [100, 200, 150],
    "Quantity" : [10, 5, 8]
}
```

```python
# Task 1:
# Create DataFrame

df = pd.DataFrame(products)
print(df)
```

```python
# Task 2:

# Print:
# shape
# columns
# data types
```

```python
shape = df.shape
print("Shape of the table =", shape)

columns = df.columns
print("Columns of the table = ", columns)

data_types = df.dtypes
print("Data types of the columns =\n", data_types)
```

```python
# ==================================================
#     Dataset for Practice (E-commerce Products)
# ==================================================
```

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"],
    "Price": [60000, 20000, 30000, 2000, 5000],
    "Quantity": [5, 10, 7, 20, 15]
}

dmart = pd.DataFrame(data)
```

```python
# Task 1 — Basic Info

# Print:
# Shape of dataset
# Column names
# Data types
```

```python
print("shape of the dataset :")
print(dmart.shape)
print("column names :\n", dmart.columns)
print("Data types:\n", dmart.dtypes)
```

```python
# Task 2 — Column Access
# Print only Price column
# Print only Product column
```

```python
print(dmart["Price"])
print(dmart["Product"])
```

```python
# Task 3 — Add Constant Column
# Category = "Electronics"
```

```python
dmart["Categorie"] = "Electronics"
dmart
```

```python
# Task 5 — Add Custom Values Column
# [10, 5, 8, 2, 6]
```

```python
dmart["Discount"] = [10, 5, 8, 2, 6]
dmart
```

```python
# Task 6 — Modify Column
# Increase all prices by 10%
```

```python
dmart["Price"] = dmart["Price"] + dmart["Price"] * 10/100
print(dmart)
```

```python
# Task 7 — Create New Derived Column
# Final_Price = Price - Discount
```

```python
dmart["Final_price"] = dmart["Price"] - dmart["Discount"]
print(dmart)
```

```python
# Task 8 — Series Practice
# Create a Series:
```

```python
ratings = pd.Series([4.5, 4.0, 4.2, 3.8, 4.3])
dmart["Ratings"] = ratings
print(dmart)
```
