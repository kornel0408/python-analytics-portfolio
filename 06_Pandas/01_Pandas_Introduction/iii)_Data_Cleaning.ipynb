```python
# *************************************************************************************************************************
#                                          DATA CLEANING WITH PANDAS 
# *************************************************************************************************************************
```

```python
data = {
    "Customer_ID": ["C1", "C2", "C3", "C4", "C5", "C5"],
    "Signup_Date": [
        "2022-01-01",
        "2022-02-15",
        "2022-03-10",
        "2022-01-01",
        "2022-05-20",
        "2022-05-20"
    ],
    "Gender": ["Male", "Female", "male", "FEMALE", "Male", None],
    "Age": [25, 30, None, 45, 28, 28],
    "Annual_Income": [50000, None, 60000, 45000, None, None],
    "City": ["Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi", "Delhi"]
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# --------------------------------
# Step-by-Step Data Cleaning
# --------------------------------
```

```python
# -----------------------
# # 1. Understand Data
# -----------------------
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
#------------------------------
# 2. Handle Missing Values
#------------------------------
```

```python
# Check missing
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
# Columns with issues:
# Age → missing
# Annual_Income → missing
# Gender → missing
```

```python
# Fix Age (Numerical → Mean)

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Age"]
df["Age"].isnull().sum()
```

```python
df["Annual_Income"] = df["Annual_Income"].fillna(df["Annual_Income"].mean())
```

```python
df["Annual_Income"]
```

```python
# Fix Gender (Categorical → Mode)

df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
# ---------------------------------------
#     3. Remove Duplicates
# ---------------------------------------
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
df.duplicated(subset=["Customer_ID"]).sum()
```

```python
#removing duplicate 
df.drop_duplicates(subset=["Customer_ID"], inplace=True)
```

```python
df
```

```python
# -------------------------------------------
#     Step 4: Fix Data Types
# -------------------------------------------
```

```python
df.info()
```

```python
# signup_date → object
# Convert to datetime

df["Signup_Date"]= pd.to_datetime(df["Signup_Date"])
df.info()
```

```python
# Age -> 31.2
#converting into proper age

# Method 1: Round → Convert to Int
df["Age"] = df["Age"].round().astype(int)
df
```

```python
# Method 2: Floor (Always Down)
df["Age"] = df["Age"].astype(int)
```

```python
# Method 3: Use np.ceil() (Always Up)
import numpy as np

df["Age"] = np.ceil(df["Age"]).astype(int)
```

```python
df["annual_income"] =df["annual_income"].round()
df
```

```python
#---------------------------------------
# 5. Clean Text Data (VERY IMPORTANT)
#---------------------------------------
```

```python
# Problem in Gender:
# Male
# male
# FEMALE
# Male (with space)
```

```python
df["Gender"] = df["Gender"].str.lower().str.strip().str.capitalize()
df
```

```python
# ---------------------------------
# 6. Standardize Column Names
# ---------------------------------
```

```python
# Customer_ID → customer_id
df.columns = df.columns.str.capitalize().str.strip()
df
```

```python
#-------------------------------------
# 7.Handling blank cells
#-------------------------------------
```

```python
# Step 1: Identify Blank Values

df.isnull().sum()
# This DOES NOT detect:" "
```

```python
# Detect empty strings
(df == "").sum()
```

```python
# Step 2: Convert ALL blanks

# import numpy as np
# df = df.replace(["", " "], np.nan)
```

```python
# Step 3: Handle Missing Values
# df["Age"] = df["Age"].fillna(df["Age"].mean())
# df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
# Drop rows
df.dropna(inplace=True)
```

```python
# Drop specific column nulls
df.dropna(subset=["Age"], inplace=True)
```
