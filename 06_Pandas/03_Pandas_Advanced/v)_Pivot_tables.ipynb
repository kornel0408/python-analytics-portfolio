```python
# What is Pivot Table?

# A pivot table is used to summarize data

# It helps you see patterns like:

# Avg salary per department
# Total sales per city
# Count of orders per product
```

```python
# Pivot = Group + Aggregate + Reshape
```

```python
import pandas as pd

data = {
    "Department": ["HR","IT","IT","HR","Finance","Finance","IT"],
    "City": ["Delhi","Mumbai","Delhi","Delhi","Chennai","Chennai","Mumbai"],
    "Salary": [50000,60000,55000,52000,80000,75000,61000]
}

df = pd.DataFrame(data)
```

```python
# -----------------------------------------
#         1.Basic Pivot Table
# -----------------------------------------
```

```python
pd.pivot_table(df, values="Salary", index="Department", aggfunc="mean")
```

```python
# -----------------------------------------
#       2.Pivot with Multiple Columns
# -----------------------------------------
```

```python
pd.pivot_table(df, values="Salary", index="Department", columns="City", aggfunc="mean")
```

```python
# -----------------------------------------
#      3.Multiple Aggregations
# -----------------------------------------
```

```python
pd.pivot_table(df, values="Salary", index="Department", aggfunc=["mean","max","min"])
```

```python
pd.pivot_table(df,values="Salary", index="Department", aggfunc={"Salary": [("average_sal", "mean"),("high", "max"),("lowest", "min")]})
```

```python
pivot = pd.pivot_table(df, values="Salary", index="Department", aggfunc=["mean","max","min"])
pivot.columns = ["average_sal", "high", "lowest"]
pivot
```

```python
# -----------------------------------------
#     4.Count Values
# -----------------------------------------
```

```python
pd.pivot_table(df, values="Salary", index="Department", aggfunc="count")
```

```python
# -----------------------------------------
#     5.Fill Missing Values
# -----------------------------------------
```

```python
pd.pivot_table(df, values="Salary", index="Department", columns="City", aggfunc="mean", fill_value=0)
```
