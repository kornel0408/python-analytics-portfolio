```python
import pandas as pd

data = {
    "Product": ["A", "B", "A", "B", "C", "A", "C"],
    "Category": ["Electronics", "Clothing", "Electronics", "Clothing", "Electronics", "Clothing", "Electronics"],
    "Sales": [100, 200, 150, 300, 250, 120, 180],
    "Quantity": [1, 3, 2, 4, 2, 1, 3]
}

df = pd.DataFrame(data)
df
```

```python
# 1. Basic GroupBy
# Total Sales per Category

df.groupby("Category")["Sales"].agg(["sum"])
```

```python
# 2. Multiple Aggregations

df.groupby("Category")["Sales"].agg(["sum", "mean", "max"])
```

```python
# 3. GroupBy Multiple Columns

df.groupby(["Category", "Product"])["Sales"].agg(["sum"])
```

```python
# 4. Count Values

df.groupby("Category")["Product"].count()
```

```python
# 5. Reset Index (Important)
#  By default, groupby makes index

df.groupby("Category")["Sales"].sum().reset_index()
```

```python
# 6. Rename Columns

df.groupby("Category")["Sales"].sum().reset_index().rename(columns={"Sales": "Total_Sales"})
```

```python
# 7. Sort Results

df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
# 8. GroupBy with Multiple Columns + Multiple Aggregations

df.groupby(["Category", "Product"]).agg({
    "Sales": "sum",
    "Quantity": "mean"
}).reset_index()
```

```python
# 9. Apply Custom Function

df.groupby("Category")["Sales"].apply(lambda x: x.max() - x.min())
```
