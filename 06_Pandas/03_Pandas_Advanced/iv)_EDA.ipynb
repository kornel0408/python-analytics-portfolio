```python
import pandas as pd

data = {
    "Order_ID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012],
    "Customer_Name": ["john","alice","bob","alice","david","eve","frank","john","grace","helen","ian","john"],
    "City": ["delhi","mumbai","delhi","mumbai","chennai","chennai","chennai","delhi","chennai","mumbai","delhi","delhi"],
    "Product": ["laptop","phone","tablet","laptop","phone","tablet","laptop","laptop","phone","tablet","laptop","phone"],
    "Quantity": [1,2,1,1,3,1,2,1,2,2,1,2],
    "Price": [50000,20000,15000,50000,22000,15000,52000,50000,22000,16000,51000,21000],
    "Order_Date": [
        "2023-01-01","2023-02-01","2023-03-01","2023-04-01",
        "2023-05-01","2023-06-01","2023-07-01","2023-08-01",
        "2023-09-01","2023-10-01","2023-11-01","2023-12-01"
    ]
}

df = pd.DataFrame(data)
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df
```

```python
# ===================================================
#         NOW: EDA (What + How + Why)
# ===================================================
```

```python
# ---------------------------------------
#     STEP 1: Basic Understanding
# ---------------------------------------
```

```python
# Number of rows → size of data
# Data types → correct or not
# Stats → mean, min, max
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# -----------------------------------------------
#         STEP 2: Add Business Column
# -----------------------------------------------
```

```python
df["Total_Amount"] = df["Quantity"] * df["Price"]
df
```

```python
# -----------------------------------------------
#         STEP 3: Total Revenue
# -----------------------------------------------
```

```python
# Insight: Total business revenue

df["Total_Amount"].sum()
```

```python
# -----------------------------------------------
#       STEP 4: Top Selling Product
# -----------------------------------------------
```

```python
# Insight:
# Which product sells most

df.groupby("Product")["Quantity"].sum().sort_values(ascending=False).reset_index()
```

```python
# -----------------------------------------------
#       STEP 5: Highest Revenue Product
# -----------------------------------------------
```

```python
# Insight:
# Which product earns most money

df.groupby("Product")["Total_Amount"].sum().sort_values(ascending=False).reset_index()
```

```python
# -----------------------------------------------
#     STEP 6: Revenue by City
# -----------------------------------------------
```

```python
# Insight:
# Best performing city

df.groupby("City")["Total_Amount"].sum().reset_index()
```

```python
# -----------------------------------------------
#   STEP 7: Top Customers
# -----------------------------------------------
```

```python
# Insight:
# Most frequent buyers

df["Customer_Name"].value_counts().reset_index(name="Number_of_purchases")
```

```python
# -----------------------------------------------
#   STEP 8: Highest Order Value
# -----------------------------------------------
```

```python
# Insight:
# Biggest single order

df.sort_values(by="Total_Amount", ascending=False).head(1)
```

```python
df.loc[df["Total_Amount"].idxmax()]
```

```python
# -----------------------------------------------
#   STEP 9: Monthly Sales Trend
# -----------------------------------------------
```

```python
# Insight:
# Which months have high sales

df.groupby("Order_Date")["Total_Amount"].sum().reset_index()
```

```python
# -----------------------------------------------
#  STEP 10: Average Order Value
# -----------------------------------------------
```

```python
# Business metric: AOV (Average Order Value)

df["Total_Amount"].mean().round()
```
