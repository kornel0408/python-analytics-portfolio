```python
#  Sample Datasets
```

```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": ["C1", "C2", "C3", "C4"],
    "Name": ["Amit", "Rahul", "Sneha", "Priya"]
})

orders = pd.DataFrame({
    "Customer_ID": ["C1", "C2", "C1", "C5"],
    "Amount": [1000, 2000, 1500, 3000]
})
```

```python
customers
```

```python
orders
```

```python
# join syntax:

# pd.merge(df1, df2, on="column", how="type")
```

```python
# 1. INNER JOIN
# Only matching values

pd.merge(customers, orders, on="Customer_ID", how="inner")
```

```python
# 2. LEFT JOIN
#All data from left table (customers)

pd.merge(customers, orders, on="Customer_ID", how="left")
```

```python
# 3. RIGHT JOIN
# All data from right table (orders)

pd.merge(customers, orders, on="Customer_ID", how="right")
```

```python
# 4. OUTER JOIN
# Everything from both tables

pd.merge(customers, orders, on="Customer_ID", how="outer")
```

```python
# Exclusive joins
```

```python
customers1 = pd.DataFrame({
    "Customer_ID": ["C1", "C2", "C3", "C4"]
})

orders1 = pd.DataFrame({
    "Customer_ID": ["C1", "C2", "C5"]
})
```

```python
# Step: Use indicator=True
```

```python
merged = pd.merge(customers, orders, on="Customer_ID", how="outer", indicator=True)
```

```python
# 1. LEFT EXCLUSIVE JOIN
# Rows only in left table
```

```python
merged[merged["_merge"] == "left_only"]
```

```python
# 2. RIGHT EXCLUSIVE JOIN
# Rows only in right table
```

```python
merged[merged["_merge"] == "right_only"]
```

```python
# 3. FULL EXCLUSIVE JOIN
# Rows NOT matching in both tables
```

```python
merged[merged["_merge"] != "both"]
```

```python
# One-Line Versions
```

```python
# Left Exclusive

pd.merge(customers, orders, on="Customer_ID", how="left", indicator=True)\
  .query('_merge == "left_only"')
```

```python
# Right Exclusive

pd.merge(customers, orders, on="Customer_ID", how="right", indicator=True)\
  .query('_merge == "right_only"')
```

```python
# Full Exclusive

pd.merge(customers, orders, on="Customer_ID", how="outer", indicator=True)\
  .query('_merge != "both"')
```

```python
# only specific columns
```

```python
# Method 1: Select Columns AFTER Merge (Most Common)
```

```python
merged = pd.merge(customers, orders, on="Customer_ID", how="inner")

result = merged[["Customer_ID", "Name", "Amount"]]

result
```

```python
# Method 2: Select Columns BEFORE Merge
```

```python
result2 = pd.merge(
    customers[["Customer_ID", "Name"]],
    orders[["Customer_ID", "Amount"]],
    on="Customer_ID",
    how="inner"
)

result2
```

```python
# Method 3: Using .loc
```

```python
merged.loc[:, ["Customer_ID", "Name", "Amount"]]
```

```python
# Method 4: Rename + Select Together
```

```python
result4 = (
    pd.merge(customers, orders, on="Customer_ID")
    [["Customer_ID", "Name", "Amount"]]
    .rename(columns={"Amount": "Total_Amount"})
)
result4
```
