```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"]
}

df = pd.DataFrame(data)
```

```python
# 1. LIKE '%Lap%' (contains)
df[df["Product"].str.contains("Lap")]
```

```python
# 2. LIKE 'Lap%' (starts with)
df[df["Product"].str.startswith("Lap")]
```

```python
# 3. LIKE '%one' (ends with)
df[df["Product"].str.endswith("ones")]
```

```python
# 4. Case Insensitive
df[df["Product"].str.contains("lap", case=False)]
```

```python
# 5. Multiple Patterns
df[df["Product"].str.contains("Lap|Mob")]
```
