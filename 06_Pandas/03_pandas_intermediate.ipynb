```python
# Selecting rows
# .loc[] (label-based)
# .iloc[] (index-based)
# Filtering (like SQL WHERE)Multiple conditions
```

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"],
    "Price": [60000, 20000, 30000, 2000, 5000],
    "Quantity": [5, 10, 7, 20, 15]
}

df = pd.DataFrame(data)
```

```python
# 1. Select Rows
```

```python
# First row
df.iloc[0]
```

```python
# Multiple rows
df.iloc[0:3]
```

```python
# 2. .iloc[] (Index Based)
```

```python
# Uses row index numbers
df.iloc[1, 2]
# Row 1, Column 2
```

```python
# Select rows & columns
df.iloc[0:3, 0:2]
```

```python
# 3. .loc[] (Label Based)
```

```python
# Uses column names
df.loc[0, "Price"]
```

```python
# Select multiple
df.loc[0:2, ["Product", "Price"]]
```

```python
# 4. Filtering
```

```python
# Products with Price > 10000
df[df["Price"] > 10000]
```

```python
# 5. Multiple Conditions
```

```python
# AND condition
df[(df["Price"] > 10000) & (df["Quantity"] > 5)]
```

```python
# OR condition
df[(df["Price"] < 5000) | (df["Quantity"] > 10)]
```

```python
# 6. Select Specific Columns After Filter
```

```python
df[df["Price"] > 10000][["Product", "Price"]]
```
