```python
# ----------------------------------------
#     Dataset 1 — E-commerce Products
# ----------------------------------------
```

```python
import pandas as pd

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch", "Monitor"],
    "Price": [60000, 20000, 30000, 2000, 5000, 15000],
    "Quantity": [5, 10, 7, 20, 15, 8]
}

df = pd.DataFrame(data)
```

```python
# Task 1 — Basic Selection
# Select first 3 rows using .iloc[]
```

```python
df.iloc[0:4]
```

```python
# Task 2 — Column Selection
# Select only Product and Price using .loc[]
```

```python
df.loc[:, ["Product", "Price"]]
```

```python
# Task 3 — Simple Filter
# Products where Price > 20000
```

```python
df[df["Price"]>20000]
```

```python
# Task 4 — Quantity Filter
# Products where Quantity >= 10
```

```python
df[df["Quantity"]>=10]
```

```python
# Task 5 — Multiple Condition (AND)
# Price > 10000 AND Quantity > 5
```

```python
df[(df["Price"]>10000) & (df["Quantity"]>5)]
```

```python
# Task 6 — Multiple Condition (OR)
# Price < 5000 OR Quantity > 10
```

```python
df[(df["Price"]<5000) | (df["Quantity"]>10)]
```

```python
# Task 7 — Column + Filter
# Show only Product and Quantity
# Where Price > 10000
```

```python
df[df["Price"]>10000][["Product", "Quantity"]]
```

```python
# Task 8 — Update with Condition
# Increase Price by 10%
# Only where Quantity > 10
```

```python
df.loc[df["Quantity"] > 10, "Price"] *= 1.10
print(df)
```

```python
# ------------------------------------------
#         Dataset 2 — Employee Data
# ------------------------------------------  
```

```python
data = {
    "Name": ["A", "B", "C", "D", "E", "F"],
    "Salary": [25000, 50000, 70000, 30000, 90000, 45000],
    "Experience": [1, 4, 6, 2, 8, 3],
    "Department": ["HR", "IT", "IT", "HR", "Finance", "IT"]
}

emp = pd.DataFrame(data)
```

```python
# Task 1 — High Salary
# Employees with Salary > 50000
```

```python
emp[emp["Salary"]>5000]
```

```python
# Task 2 — Department Filter
# Employees in "IT" department
```

```python
emp[emp["Department"].str.contains("IT")]
```

```python
# Task 3 — Multiple Condition
# Salary > 40000 AND Experience > 3
```

```python
emp[(emp["Salary"]>40000) & (emp["Experience"]>3)]
```

```python
# Task 4 — OR Condition
# Department = HR OR Salary > 60000
```

```python
emp[(emp["Department"].str.contains("HR")) | (emp["Salary"]>60000)]
```

```python
# # Task 5 — LIKE Operator
# Names containing "A"
```

```python
emp[emp["Name"].str.contains("A")]
```

```python
# Task 6 — Column + Filter
# Show only Name and Salary
# Where Experience >= 3
```

```python
emp[emp["Experience"]>3][["Name", "Salary"]]
```

```python
# Task 7 — Update Data 
# Increase salary by 5000
# Only for IT employees
```

```python
emp.loc[emp["Department"] == "IT", "Salary"] += 5000

print(emp)
```

```python
# Task 8 — Advanced Filter 
# Employees:
# Experience between 3 and 6
```

```python
emp[(emp["Experience"]>=3) & (emp["Experience"]<=6)]
```

```python
# ---------------------------------------
#     Dataset 3 — Student Data
# ---------------------------------------
```

```python
data = {
    "Name": ["John", "Sam", "Anna", "Mike", "Sara"],
    "Math": [78, 40, 88, 60, 30],
    "Science": [85, 35, 92, 65, 25],
    "English": [90, 50, 95, 70, 40]
}

students = pd.DataFrame(data)
```

```python
# Task 1 — Failed Students
# Students where any subject < 35
```

```python
failed_students = students[
    (students[["Math", "Science", "English"]] < 35).any(axis=1)
]

failed_students
```

```python
# Task 2 — Top Students
# Students where all subjects > 80
```

```python
Top_students = students[
    (students[["Math", "Science", "English"]]>80).all(axis=1)
]
Top_students
```

```python
# Task 3 — Average Filter
# Students with average > 70
```

```python
avg = students[["Math", "Science", "English"]].mean(axis=1)
result = students[avg > 70]
result
```

```python
# Task 4 — Column Selection + Filter
# Show Name and Math
# Where Science > 60
```

```python
students[students["Science"]>60] [["Name", "Math"]]
```

```python
# Task 5 — Update Marks 🔥
# Add 5 marks
# Only where average < 50
```

```python

avg = students[["Math", "Science", "English"]].mean(axis=1)
condition = avg < 50
students.loc[condition, ["Math", "Science", "English"]] += 5
print(students)
```
